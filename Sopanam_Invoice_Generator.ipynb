```python
from io import StringIO, BytesIO
from docx.shared import Cm
from docxtpl import DocxTemplate, InlineImage
from docx.shared import Cm, Inches, Mm, Emu

from datetime import date, timedelta
import json
import os
```

```python
def get_context():
    """ You can generate your context separately since you may deal with a lot 
        of documents. You can carry out computations, etc in here and make the
        context look like the sample below.
    """
    # input
    
    #invoice_items = 3 #int(input("Number of tasks in the invoice being created = "))
    item = 1
    add_item = True
    invoice_items = []
    subtotal = 0.00
    tax_rate = 0.13
    tax_amt = subtotal*tax_rate
    amt_paid = 0.00
    total = subtotal+tax_amt
    amt_due = total-amt_paid
    
    while add_item:
        print("Enter task " + str(item) + " details for the invoice being created:")
        description = input("Description of task = ")
        quantity = float(input("Number of hours to perform task = "))
        rate = float(input("Hourly rate of performing task = "))
        amount = quantity*rate
        subtotal = subtotal + amount
        print("Amount = "+str(amount))
        print(" ")
        
        invoice_items.append({'description': description, 'quantity': quantity, 'rate': rate, 'amount':amount})
        
        add = input("Would you like to enter another task to the current invoice? (Y/N): ")
        if add == "Y" or add == "y":
            add_item = True
            item = item + 1
        else:
            add_item = False    
    
    tax_amt = subtotal*tax_rate
    total = subtotal+tax_amt
    amt_due = total - amt_paid
    
    print("Invoice details....")
    print("------------------------------------------------")
    print("Subtotal = " + str(subtotal))
    print("HST = " + str(tax_amt))
    print("Total = " + str(total))
    print("Amount paid = " + str(amt_paid))
    print("Amount due = " + str(amt_due))
    
    return {
        'invoice_no': 12345,
        'date': date.today().isoformat(),
        'due_date': (date.today()+timedelta(days=30)).isoformat(),
        'name': name,
        'address': address,
        'subtotal': subtotal,
        'tax_amt': tax_amt,
        'total': total,
        'amt_paid': amt_paid,
        'amt_due': amt_due,
        'row_contents': invoice_items
    }
```

```python
def from_template(template, signature):

    template = DocxTemplate(template)
    context = get_context() # gets the context used to render the document
    print("Adding Signature to the invoice.....")
    img_size = Cm(7)  # sets the size of the image
    sign = InlineImage(template, signature, img_size)
    context['signature'] = sign  # adds the InlineImage object to the context

    target_file = BytesIO()
    template.render(context)
    template.save(target_file)
    print("Generating the invoice.....")

    return target_file
```

```python
template = 'InvoiceTemplate.docx'
signature = 'signature.png'

#input
print("Enter invoice details....:")
name = input("Name of person or org to be invoiced = ")
address = input("Address of person or org to be invoiced = ")

with open("Invoice"+"_"+name+"_"+date.today().isoformat()+".docx", "wb") as outfile:
    # Copy the BytesIO stream to the output file
    outfile.write(from_template(template, signature).getbuffer())
    
print("Invoice generation complete.....")
print("=============================================================================")
    
```
